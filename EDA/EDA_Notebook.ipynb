```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Leer el dataset y ver que columnas tiene

```python

df = pd.read_csv("test.csv")
print("Dimensiones del dataset:", df.shape)
```

```python
print("\nColumnas del dataset:")
print(df.columns.tolist())
```

Tipos de respuesta por cada columna

```python
for col in df.columns:
    print(f"\nColumna: {col}")
    print(df[col].unique())
```

```python
print("\nPrimeras filas del dataset:")
display(df.head())
```

Porcentaje de valores


```python
print("Valores nulos por columna:")
print(df.isnull().sum())

porcentaje_nulos = (df.isnull().sum() / len(df)) * 100
print("Porcentaje de valores nulos:")
print(porcentaje_nulos)
```

Porcentaje de ceros

```python
porcentaje_ceros = (df.isin([0]).sum() / len(df)) * 100
print(porcentaje_ceros)
```

4. Detección de outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = df.select_dtypes(include=["int64", "float64"]).columns

for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col], color="skyblue")
    plt.title(f"Diagrama de caja y bigotes - {col}")
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

5. Exploracion de cada variable mediante histogramas

```python
# Correlaciones
plt.figure(figsize=(10,8))
sns.heatmap(df[num_cols].corr(), cmap="coolwarm", annot=False)
plt.title("Matriz de correlación")
plt.show()

# Histogramas de distribución
df[num_cols].hist(figsize=(14,10), bins=30)
plt.suptitle("Distribución de variables numéricas")
plt.show()

# Conteos de variables categóricas
cat_cols = df.select_dtypes(include=["object"]).columns
for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=df)
    plt.title(f"Distribución de {col}")
    plt.xticks(rotation=30)
    plt.show()
```

# **¿Qué hacer con los nulos?**
Dado que en nuestra base de datos no hay exisencia de datos nulos, no es necesario plantearse ninguna estrategia para tratarlos.

# **¿Qué hacer con los ceros?**
Los ceros que existen en nuestra base de datos son respuestas binarias de preguntas de si o no. O también calificaciones, por lo que no podemos la libertad de eliminarlos o modificarlos.
